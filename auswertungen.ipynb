# Auswerten der Lochplatte

```python
import numpy as np
import matplotlib.pyplot as plt
import json
```

```python
pfad_res = np.loadtxt('platte-pfad-s1.dat',delimiter=',')
```

```python
fig, ax = plt.subplots(ncols=1, nrows=1,figsize=np.array((12,10))/2.54)

ax.set_xlabel('x-Position $x$ (mm)')
ax.set_ylabel('max. princ. stress $\sigma_I$')

ax.plot(pfad_res[:,0],pfad_res[:,1],'o-')

plt.tight_layout()
plt.savefig('pfad-auswertung.png',dpi=600)

plt.show()
```

```python
# Diagramm vom CheatSheet
eps = np.linspace(0,1)
sigma = eps**2
fig, ax = plt.subplots(ncols=1, nrows=1,figsize=np.array((8.,5.))/2.54)
# realstring for special characters / latex formulas
ax.set_title(r'stress $sigma$ over strain $varepsilon$', fontsize=10)
# plot a curve
ax.plot(eps, sigma, 'r', label='exp.')

# axis labels
ax.set_xlabel(r'eng. strain $varepsilon_ mathrm{t}$')
ax.set_ylabel(r'eng. stress $sigma_ mathrm{t}$')
# change the axis limits
ax.set_xlim(0,1)
ax.set_ylim(ymin=0)
# display a legend
leg = ax.legend(ncol=1)
# set a legend title
leg.set_title('legend title', prop={'size':12})
# show a grid
ax.grid()
# fit diagram into image size
fig.tight_layout(pad=0)
# save as .png (dpi:resolution) or .pdf
fig.savefig('sig_eps.png', dpi=300)
fig.show() # display plot in Python
```
